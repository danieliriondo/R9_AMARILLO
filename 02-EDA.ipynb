```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calcular rendimientos y estadísticas
rendimientos = df.pct_change().dropna()
volatilidades = rendimientos.std() * np.sqrt(252)
rendimientos_anuales = rendimientos.mean() * 252

# 2. Matriz de correlaciones
correlaciones = rendimientos.corr()

# 3. Visualización
plt.figure(figsize=(15, 12))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlaciones - Dow Jones')
plt.show()

# 4. Ranking por diferentes criterios
criterios = pd.DataFrame({
    'Volatilidad': volatilidades,
    'Rendimiento': rendimientos_anuales,
    'Sharpe_Ratio': rendimientos_anuales / volatilidades
}).round(4)

print(criterios.sort_values('Sharpe_Ratio', ascending=False))
```
